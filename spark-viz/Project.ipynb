```
place_rating = place.join(rating,"placeID")
display(place_rating)
```

```
# show the schema of the dataframe
place_rating.printSchema()
```

```
#Show the dataset columns
place_rating.columns
```

```
import pyspark.sql.functions as f
#Most listed and visited restaurant
restaurant = (
  place_rating
  .groupBy(f.col("name").alias("restaurant"))
  .agg(f.count("name").alias("count"))
  .sort(f.desc("count"))
  
)
display(restaurant)
```

```
import pyspark.sql.functions as f
#City with the most represented restaurant
city = (
  place_rating
  .groupBy(f.col("city").alias("City"))
  .agg(f.count("city").alias("count"))
  .sort(f.desc("count"))
  
)
display(city)
```

```
import pyspark.sql.functions as f
#The diferent type of restaurant and their proportion
rest_type = (
  place_rating
  .groupBy(f.col("Rambience").alias("Type of ambience"))
  .agg(f.count("Rambience").alias("count"))
  .sort(f.desc("count"))
  
)
display(rest_type)
```

```
import pyspark.sql.functions as f
#The rating of the restaurant
rating_note = (
  place_rating
  .groupBy(f.col("rating").alias("rating note"))
  .agg(f.count("rating").alias("count"))
  .sort(f.desc("count"))
  
)
display(rating_note)
```

```
import pyspark.sql.functions as f
display(
  place_rating.select("name", "other_services", "price")
             .filter(f.col("other_services") == "internet")
)
```

```
from pyspark.sql.types import * 
import pyspark.sql.functions as f

def length(string):
  return 2*len(string)

lengthUDF = f.udf(length, IntegerType())

display(
  place_rating.select(
    f.col("name"),
    lengthUDF(f.col("name")) # équivalent pandas : apple_store['track_name'].apply(length)
  ) 
)
```

```
import pyspark.sql.functions as f
mean_ratings = place_rating.groupBy('placeID').agg(f.mean('rating').alias('mean_rating'),f.count('rating').alias('number_ratings'))
display(mean_ratings)
```

```
display(mean_ratings)
```

```
import pyspark.sql.functions as f
predictions = mean_ratings.withColumn("mean_rating", f.abs(f.round(mean_ratings["mean_rating"],0)))
display(predictions)
```

```
from pyspark.sql.types import *
from pyspark.sql import functions as F

xRate = mean_ratings.select(
  F.col("placeID").cast(IntegerType()),
  F.col("mean_rating").cast(IntegerType()),
  F.col("number_ratings").cast(IntegerType())
)
```
